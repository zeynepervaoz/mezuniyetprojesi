```python
import requests

url = 'https://www.mta.gov.tr/v3.0/bilgi-merkezi/2010-islenmis-mermer-ihracati' # Hedef web sitesi URL'sini alyoruz.
response = requests.get(url)

# Web sitesinden ierii seiyoruz.
content = response.content

# erii grntyoruz.
print(content)
```

```python
from bs4 import BeautifulSoup

# BeautifulSoup kullanarak ierii analiz ediyoruz.
soup = BeautifulSoup(content, 'html.parser')

# rnein, <kilogram> etiketlerini ekmek istediimizde <soup.find_all> kodunu kullanyoruz.
paragraphs = soup.find_all('td')

# ekilen metni <print> ile grntlyoruz.
for td  in paragraphs:
    print( td.text)
```
