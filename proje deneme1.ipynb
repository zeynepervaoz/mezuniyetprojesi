```python
import requests

url = 'https://www.mta.gov.tr/v3.0/bilgi-merkezi/2010-islenmis-mermer-ihracati' # Hedef web sitesi URL'sini alıyoruz.
response = requests.get(url)

# Web sitesinden içeriği seçiyoruz.
content = response.content

# İçeriği görüntüyoruz.
print(content)
```

```python
from bs4 import BeautifulSoup

# BeautifulSoup kullanarak içeriği analiz ediyoruz.
soup = BeautifulSoup(content, 'html.parser')

# Örneğin, <kilogram> etiketlerini çekmek istediğimizde <soup.find_all> kodunu kullanıyoruz.
paragraphs = soup.find_all('td')

# Çekilen metni <print> ile görüntülüyoruz.
for td  in paragraphs:
    print( td.text)
```
